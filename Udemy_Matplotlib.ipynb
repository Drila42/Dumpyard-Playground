```
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
x=[1,2,3,4]
y=[1500,1400,670,3456]
plt.plot(x,y)
legend=["jan","feb","march","apr"]
plt.xticks(x,legend)
plt.show()
```

```
plt.bar(x,y)
```

```
plt.bar(x,y)
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.show()
```
